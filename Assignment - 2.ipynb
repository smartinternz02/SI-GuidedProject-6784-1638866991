# Deal with Missing values, Outliers and remove unwanted columns

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
data.drop("deck",axis=1 ,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.reset_index(drop=False,inplace = True)
```

```python
mv=data[data["embarked"].isnull()].index.tolist()
```

```python
data.drop(mv, axis=0, inplace=True)
```

```python
data.isnull().any()
```

```python
data.reset_index(drop=False,inplace = True)
data
```

```python
data["age"].median()
```

```python
data["age"].describe()
```

```python
data.fillna(data.median(), inplace= True)
```

```python
data.isnull().any()
```

```python
sns.boxplot(data["age"])
```

```python
q1=data["age"].describe()["25%"]
q3=data["age"].describe()["75%"]
iqr=q3-q1
l_b=q1 -(1.5 *iqr)
u_b=q3 +(1.5*iqr)
print(l_b)
print(u_b)
```

```python
(data[data["age"]<l_b].shape[0]/data["age"].shape[0])*100
```

```python
(data[data["age"]>u_b].shape[0]/data["age"].shape[0])*100
```

```python
data.drop(data[data["age"]<l_b].index, axis=0, inplace=True)
```
