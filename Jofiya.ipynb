```python
def speed(s,b):
    if s<=60:
        print("NoTicket")
    elif s>60 and s<=80:
        print("Small Ticket")
    elif s>80:
        print("Big Ticket")

s=int(input("Enter speed"))
b=bool(input("Is it your birthday? (0/1)"))
if(b==1):
    s=s-5
    speed(s,b)
else:
    speed(s,b)
```
